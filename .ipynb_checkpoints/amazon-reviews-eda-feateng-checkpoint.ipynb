- highly imbalanced data set

```python
def get_avg_words(texts):
    words = []
    for text in texts:
        _words = text.split()
        words.extend(_words)
    return len(words)/len(texts)
```

```python
print('Average words perREVIEW Title in train:', get_avg_words(train_df['Review Title']))
print('Average words per REVIEW Title in test:',get_avg_words(test_df['Review Title']))
```

```python
print('Average words per REVIEW Text in train:', get_avg_words(train_df['Review Text']))
print('Average words per REVIEW Text in test:',get_avg_words(test_df['Review Text']))
```

- both train and test are having almost same word distribution, test being a little more in no.of words in REVIEW TECT while train having no.of words in REVIEW TITLE.

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
stopwords = set(STOPWORDS)

def draw_wordcloud(data, title = None):
    wordcloud = WordCloud(
        background_color='black',
        stopwords=stopwords,
        max_words=200,
        max_font_size=40, 
        scale=3,
        random_state=1 # chosen at random by flipping a coin; it was heads
    ).generate(str(data))

    fig = plt.figure(1, figsize=(9,9))
    plt.axis('off')
    if title: 
        fig.suptitle(title, fontsize=20)
        fig.subplots_adjust(top=2.3)

    plt.imshow(wordcloud)
    plt.show()
```

```python
train_df['text'] = train_df['Review Title'] + ' ' + train_df['Review Text']
test_df['text'] = test_df['Review Title'] + ' ' + test_df['Review Text']
```

```python
draw_wordcloud(train_df['text'], 'Most Common words in Train Corpus')
draw_wordcloud(test_df['text'], 'Most Common words in Test Corpus')
```

- There is clear overlapping of words in Train and Test Data which is very much desirable.

```python
for value in train_df['topic'].unique():
    sample_df = train_df[train_df['topic'] == value]
    draw_wordcloud(sample_df['text'], 'Most Common words among ""{}"" data points'.format(value))
```

### Feat Eng

```python
import string
def apply_count_features(df):
    punctuation=string.punctuation
    df['word_count']=df['text'].apply(lambda x: len(str(x).split(" ")))
    df['char_count'] = df['text'].str.len()
    
    def avg_word(sentence):
        words = sentence.split()
        return (sum(len(word) for word in words)/len(words))

    df['avg_word'] = df['text'].apply(lambda x: avg_word(x))
    from nltk.corpus import stopwords
    stop = stopwords.words('english')

    df['stopwords'] = df['text'].apply(lambda x: len([x for x in x.split() if x in stop]))
    df['numerics'] = df['text'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
    
    df['word_density'] = df['char_count'] / (df['word_count']+1)
    df['punctuation_count'] = df['text'].apply(lambda x: len("".join(_ for _ in x if _ in punctuation))) 
    return df
```

```python
train_df = apply_count_features(train_df)
test_df = apply_count_features(test_df)
```

```python
train_df.groupby('topic').describe().transpose()
```

- avg_words and word_density are almost equal for all classes.
- punctuation count, stop words and char counts seems to differentiate a b/w the classes we going to predict.

### As a multi-label problem

- As there are duplicate texts with different labels (in both train and test), we group those and formulate this as multi-label classification problem.

```python
train_multi_label = train_df.groupby('text')['topic'].apply(lambda x: '|'.join(x)).reset_index()
print('Total combinations of multi-labels in the train using the 21 classes:', 
      len(train_multi_label['topic'].value_counts().to_dict()))
```

```python
test_multi_label = test_df.groupby('text').count().reset_index()
test_multi_label['review_count'] = test_multi_label['Review Text']
```
